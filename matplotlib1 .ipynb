```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#QUESTION 2
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y= 3+ np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
# QUESTION 3
'''subplot() function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time'''
x=np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
# QUESTION 4
"Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point, and the width of each bar represents the category of the data. "
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
# QUESTION 5
"A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. "
box1 = np.random.normal(100, 10, 200)
plt.subplot(2,2,1)
plt.boxplot(box1)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,2,2)
plt.boxplot(box2)
```
